```python
from detection.Form import *
from detection.Image import Image
import cv2
import numpy as np
import json
from glob import glob
from os.path import join as pjoin
import os
import time
```

```python
data_root = 'E:\\Mulong\\Datasets\\form\\input'
export_root = 'E:\\Mulong\\Result\\form'
form_imgs = pjoin(data_root, '*.jpg')
```

```python
times = []
bad_img = 0
for i, form_img in enumerate(glob(form_imgs)):
    start = time.clock()
    print('Form number:%d, Bad Image:%d, Form Image:%s' %(i, bad_img, form_img))
    
    try:
        form_compo_detection(form_img, export_dir=export_root)
        times.append(round(time.clock() - start, 3))
        print("Time taken:%.3fs" % (time.clock() - start), '\n\n')
    except:
        bad_img += 1
        print('*** Bad Img %d:%s ***' %(bad_img, form_img), '\n\n')
```

```python
f = open(pjoin(export_root, 'runtime.txt'), 'w')
for t in times:
    f.write(str(t) + '\n')
f.close()
```
